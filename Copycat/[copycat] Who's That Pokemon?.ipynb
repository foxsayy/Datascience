## 3단계: 아무 이미지 넣기
- 포켓몬 이미지가 256 x 256 이미지라 비교를 위해 넣어주는 이미지도 같은 차원의 이미지여야 한다.
- 크기가 다른 이미지의 경우 Pillow의 `resizeimage`를 사용하면 된다.

## 4단계: 코드 작성

```python
import numpy as np
from scipy.spatial import distance
from PIL import Image
from resizeimage import resizeimage
```

```python
paths = ['./pokemon/' + str(i) + '.png' for i in range(1, 386)]
pokepics = [imread(path) for path in paths]
```

```python
def resize(file):
    with open(file, 'r+b') as f:
        with Image.open(f) as original_image:
            return np.array(original_image.resize([256, 256]))

def displayboth(image, poke):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))
    for ax, title in zip([ax1, ax2], ['Input', 'BestMatch']):
        ax1.imshow(image)
        ax2.imshow(pokepics[poke-1])
        ax.set_title(title)
    plt.show()

def imgdistance(image, poke):
    first, second = image, pokepics[poke-1]
    dist = 0
    for i in range(256):
        for j in range(256):
            dist += distance.euclidean(first[i][j], second[i][j])
    return np.mean(dist)

def imgdistlist(image):
    return [(imgdistance(image, i+1), i+1) for i in range(151)]

def bestfitpokemon(pic):
    image = resize(pic)
    distances = imgdistlist(image)
    return displayboth(image, (sorted(distances)[1][1]))
```
